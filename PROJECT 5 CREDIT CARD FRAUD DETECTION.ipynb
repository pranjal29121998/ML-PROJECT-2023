# undersampling 

```python


normal = df[df['Class']==0]
fraud = df[df['Class']==1]
```

```python
normal.shape[0]
```

```python
fraud.shape[0]
```

```python
normal_sample = normal.sample(n=473)
```

```python
normal_sample.shape[0]
```

```python
new_data = pd.concat([normal_sample,fraud])
```

```python
new_data.shape
```

```python
new_data['Class'].value_counts()
```

```python
X = new_data.drop(columns='Class')
y = new_data['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.20,random_state=42)
X_train.shape[0],X_test.shape[0],y_train.shape[0],y_test.shape[0]
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred_1 = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report
```

```python
accuracy_score(y_test,y_pred_1)
```

```python
precision_score(y_test,y_pred_1)
```

```python
recall_score(y_test,y_pred_1)
```

```python
f1_score(y_test,y_pred_1)
```

```python
cr = classification_report(y_test,y_pred_1)
print(cr)
```

# Decision Tree Classifier

```python
y_pred_2 = dt.predict(X_test)
```

# Over Sampling

```python
df.columns
```

```python
X= df.drop(columns='Class')
y = df['Class']
```

```python
from imblearn.over_sampling import SMOTE
X_res,y_res = SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_res,y_res, test_size=0.20, random_state=42)
X_train.shape[0] ,X_test.shape[0] ,y_train.shape[0] ,y_test.shape[0] 
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```
